# Visualization using Matplotlib

## Importing Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

## Reading Data

```python
house_data = pd.read_csv('../house_price.csv')
house_data.head()
```

### Scatter plot with Regression line

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='Size', y='Price', data=house_data, scatter_kws={'color':'blue'}, line_kws={'color':'red'})
plt.title('Scatter Plot with Regression Line: Size vs Price')
plt.xlabel('Size (sq ft)')
plt.ylabel('Price (in $1000)')
plt.grid(True)
plt.show()
```

### Heatmap

```python
corr = house_data.corr()

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

### Histogram of House Prices

```python
plt.figure(figsize=(10, 6))
sns.histplot(house_data['Price'], bins=20, color='skyblue', kde=True)
plt.title('Distribution of House Prices')
plt.xlabel('Price (in $1000)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Bar Plot of Average Price by Number of Bedrooms

```python
# Calculate average price by number of bedrooms
avg_price_bedrooms = house_data.groupby('Bedrooms')['Price'].mean().reset_index()

# Bar Plot of Average Price by Number of Bedrooms
plt.figure(figsize=(10, 6))
sns.barplot(x='Bedrooms', y='Price', data=avg_price_bedrooms, hue = 'Bedrooms')
plt.title('Average Price by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Average Price (in $1000)')
plt.grid(True)
plt.show()
```

### Box Plot 

```python
# Box Plot of Price by Number of Bedrooms
plt.figure(figsize=(10, 6))
sns.boxplot(x='Bedrooms', y='Price', data=house_data, hue = 'Bedrooms', legend = False)
plt.title('Price Distribution by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price (in $1000)')
plt.grid(True)
plt.show()
```

### Pair Plot

```python
plt.figure(figsize=(12, 10))
sns.pairplot(house_data, vars=['Size', 'Bedrooms', 'Age', 'Price'], diag_kind='kde', plot_kws={'alpha':0.5})
plt.suptitle('Pair Plot of House Data', y=1.02)
plt.show()
```

### Combining all Subplots 

```python
# Compute correlation matrix
corr = house_data.corr()

# Create subplots
fig, axs = plt.subplots(3, 2, figsize=(18, 18))

# Scatter Plot with Regression Line
sns.regplot(x='Size', y='Price', data=house_data, scatter_kws={'color':'blue'}, line_kws={'color':'red'}, ax=axs[0, 0])
axs[0, 0].set_title('Scatter Plot with Regression Line: Size vs Price')

# Histogram of House Prices
sns.histplot(house_data['Price'], bins=20, color='skyblue', kde=True, ax=axs[0, 1])
axs[0, 1].set_title('Distribution of House Prices')

# Bar Plot of Average Price by Number of Bedrooms
avg_price_bedrooms = house_data.groupby('Bedrooms')['Price'].mean().reset_index()
sns.barplot(x='Bedrooms', y='Price', data=avg_price_bedrooms, hue = 'Bedrooms', ax=axs[1, 0])
axs[1, 0].set_title('Average Price by Number of Bedrooms')

# Box Plot of Price by Number of Bedrooms
sns.boxplot(x='Bedrooms', y='Price', data=house_data, hue = 'Bedrooms', ax=axs[1, 1])
axs[1, 1].set_title('Price Distribution by Number of Bedrooms')

# Heatmap of Correlation Matrix
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', ax=axs[2, 0], vmin=-1, vmax=1)
axs[2, 0].set_title('Heatmap of Correlation Matrix')

# Empty plot to maintain layout
axs[2, 1].axis('off')

# Adjust layout
plt.tight_layout()
plt.show()
```
