# Visualization using Matplotlib

## Importing Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Reading Data

```python
house_data = pd.read_csv('../house_price.csv')
house_data.head()
```

### Scatter Plot

```python
# Scatter plot for Size vs Price
plt.figure(figsize=(10, 6))
plt.scatter(house_data['Size'], house_data['Price'], alpha=0.5)
plt.title('Size vs Price')
plt.xlabel('Size (sq ft)')
plt.ylabel('Price (in $1000)')
plt.grid(True)
plt.show()
```

### Histogram 

```python
# Histogram for the distribution of house prices
plt.figure(figsize=(10, 6))
plt.hist(house_data['Price'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of House Prices')
plt.xlabel('Price (in $1000)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Bar Plot

```python
# Bar plot for the average price by the number of bedrooms
avg_price_bedrooms = house_data.groupby('Bedrooms')['Price'].mean()
plt.figure(figsize=(10, 6))
avg_price_bedrooms.plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title('Average Price by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Average Price (in $1000)')
plt.grid(True)
plt.show()
```

### Difference between hist plot and bar plot

```python
# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(16, 6))

# Histogram: Distribution of House Prices
axs[0].hist(house_data['Price'], bins=20, color='skyblue', edgecolor='black')
axs[0].set_title('Distribution of House Prices')
axs[0].set_xlabel('Price (in $1000)')
axs[0].set_ylabel('Frequency')
axs[0].grid(True)

# Bar Plot: Average Price by Number of Bedrooms
avg_price_bedrooms = house_data.groupby('Bedrooms')['Price'].mean()
avg_price_bedrooms.plot(kind='bar', color='lightcoral', edgecolor='black', ax=axs[1])
axs[1].set_title('Average Price by Number of Bedrooms')
axs[1].set_xlabel('Number of Bedrooms')
axs[1].set_ylabel('Average Price (in $1000)')
axs[1].grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```

### Box Plot

```python
# Box plot for Price by number of bedrooms
plt.figure(figsize=(10, 6))
house_data.boxplot(column='Price', by='Bedrooms', grid=False)
plt.title('Price Distribution by Number of Bedrooms')
plt.suptitle('')  # Suppress the automatic 'Boxplot grouped by Bedrooms' title
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price (in $1000)')
plt.grid(True)
plt.show()
```

### Line Plot

```python
# Sample Data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Line Plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, color='b', linestyle='-', marker='o')
plt.title('Line Plot of Sine Function')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True)
plt.show()
```

### Subplots with different plot types

```python
# Sample Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Subplots with Different Plot Types
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Line Plot
axs[0, 0].plot(x, y1, 'r-')
axs[0, 0].set_title('Line Plot')
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('Y')

# Scatter Plot
axs[0, 1].scatter(x, y1, color='b', marker='o')
axs[0, 1].set_title('Scatter Plot')
axs[0, 1].set_xlabel('X')
axs[0, 1].set_ylabel('Y')

# Histogram
axs[1, 0].hist(y1, bins=20, color='purple', edgecolor='black')
axs[1, 0].set_title('Histogram')
axs[1, 0].set_xlabel('Values')
axs[1, 0].set_ylabel('Frequency')

# Pie Chart
sizes = [20, 30, 25, 25]
labels = ['A', 'B', 'C', 'D']
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen'])
axs[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()
plt.show()
```
